```python
from ift6758.data import get_player_stats
import pandas as pd
import matplotlib.pyplot as plot
```

```python
# pandas settings: display all columns
pd.set_option("display.max_columns", None)
```

```python
# Retrieve players stats
df = get_player_stats(2017, 'goalies')
df.head()
```

```python
# Sort goalies by their save percentage SV%
df['SV%'] = df['SV%'].apply(pd.to_numeric)
df_sorted = df.sort_values('SV%', ascending=False)
df_sorted.head()
```

```python
# Sort goalies by their GSAA (Goals Saved Above Average)
df['GSAA'] = df['GSAA'].apply(pd.to_numeric)
df_sorted2 = df.sort_values('GSAA', ascending=False)
df_sorted2.head(10)
```

```python
df_sorted2.head(20).plot.barh(y='SV%', x="Player", title="Performance of the top 20 goalies for the 2017-2018 season")
plot.show(block=True)
```

```python
df_sorted2.head(20).plot.barh(y='GSAA', x="Player", title="Performance of the top 20 goalies for the 2017-2018 season")
plot.show(block=True)
```
