```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('dataset/season2020.csv')
```

```python
df_mtl = df.loc[df['teamOfShooter'] == 'Montr√©al Canadiens']
```

```python
data = df_mtl.loc[df_mtl['eventType'] == 'Goal']
```

```python
fig, ax = plt.subplots(figsize=(13,8.5))
plt.axis('off')

img = plt.imread('../figures/nhl_rink.png')

# rink dimensions
ext = [-100, 100, -42.5, 42.5]
plt.imshow(img, zorder=0, extent=ext, aspect=1)

#plot the points on top of the image
plt.scatter(data['xCoord'],data['yCoord'], s=100,c='#ea6969',alpha=.7)
```
